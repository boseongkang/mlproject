# 1. Making our own billboard song dictionary

```python
chart = b.ChartData('hot-100')
chart.title
```

```python
song = chart[1]
song.title
```

```python
song_lib = []

for song in chart:
    song_lib.append({
        'rank': song.rank,
        'title': song.title,
        'artist': song.artist
    })
```
